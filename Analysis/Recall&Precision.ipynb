```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
def RecallPrecisionPlot(df, name):
    
    plt.figure(figsize=(20,8))

  #  colors = ['green','yellow','red','blue']
    new_df = df[df["Classifier"].str.contains(name)] 
    ## comment for multiple
  #  new_df2 =(df[df["Classifier"].str.contains('M')])
  #  new_df = new_df.append(new_df2)    
    df = new_df.reset_index()
    x = []
    y = []
    x1 = ''
    y1 = 0
    x11 = []
    y11 = []
    y12 = 0
    for t in(df['Classifier']):
        if 'MP' in t or 'SP' in t or 'MT' in t or 'ST' in t:
            x.append(t)
            df_new = df[df["Classifier"].str.contains(t)]
            y.append(df_new['sens'].values)
            
            x11.append(t)
            df_new = df[df["Classifier"].str.contains(t)]
            y11.append(df_new['prec'].values)
        else:
            x1 = t
            
            df_new = df[df["Classifier"].str.contains(t)]
            y1 = df_new['sens'].values
            y12 = df_new['prec'].values



    y1 = [y1[0]]* (len(df)-1)
    y12 = [y12[0]]* (len(df)-1)



    plt.plot(x, y, color='green', marker='o', linestyle='dashed',linewidth=2, markersize=15, label='Recall of Reclassified')
    
    plt.plot(x, y1, color='red',linewidth=2, markersize=15, label=name+' Recall')

    plt.plot(x11, y11, color='blue', marker='o', linestyle='dashed',linewidth=2, markersize=15, label='Precision of Reclassified')
    
    plt.plot(x, y12, color='red',linewidth=2, markersize=15, label=name+' Precision')


    
    plt.ylim(0,1.05)
    plt.yticks(np.arange(0,1.05,0.05))
    plt.grid(True)
    plt.legend(loc='lower right')
  #  plt.xticks(rotation=45)


    plt.xlabel('Type of reassignment',fontsize=14)
  #  plt.ylabel('Recal [0.0 - 1.0]',fontsize=14)
    
    plt.title('Recall values',fontweight='bold',fontsize=20)
```

```python
def ClassifiersPlt22(df,name):
    new_df = df[df["Classifier"].str.contains(name)] 
 #   new_df2 =(df[df["Classifier"].str.contains('M')])
 #   new_df = new_df.append(new_df2)    
    new_df = new_df.reset_index()
    
    
    ##colors
    #C1CF79   verde 
    #C57673   marroncino
    #B375CC   viola
    #7797CD   blue
    #87CE9B   verde acceso
    
    label = new_df['Classifier']
    y1 = new_df['sens']
    y2 = new_df['prec']
    
    x = np.arange(len(label))  # the label locations
    width = 0.20  # the width of the bars

  #  plt.figure(figsize=(15,5))
    fig, ax = plt.subplots(figsize=(20,10))
    
    rects1 = plt.bar(x - width/2, y1, width, label='Recall', color='#87CE9B')
    rects2 = plt.bar(x + width/2, y2, width, label='Precision', color='#B375CC')

    # Add some text for labels, title and custom x-axis tick labels, etc.
    plt.ylabel('Scores')
   # plt.title('Scores by group and gender')
    plt.xticks(x)
    ax.set_xticklabels(label)
  #  plt.legend()


    plt.hlines(y = y1[0], xmin = x[0]-width, xmax = x[-1]+width, color = 'blue', label='Original Recall')
    
    plt.hlines(y = y2[0], xmin = x[0], xmax = x[-1]+width, color = 'red', label = 'Original Precision' )

    plt.legend(loc='best')


    #   ax.bar_label(rects1, padding=3)
 #   ax.bar_label(rects2, padding=3)

  #  fig.tight_layout()
    plt.savefig('recall-prec-'+name+'-species')


    plt.show()
```
