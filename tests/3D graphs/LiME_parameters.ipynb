```python
# libraries
import numpy as np
import matplotlib.pyplot as plt
 
# set width of bars
barWidth = 0.25
 
# set heights of bars
bars1 = [
64628,
78263,
143334,
209815,
212780,
361316,
665759,
1067713,
1701232,
1203729,
2221942
]

bars2 = [
1314601,
1182805,
796125,
627607,
655077,
318142,
62446,
62267,
61606,
61956,
55441
]

plt.figure(figsize=(15,10))

# Set position of bar on X axis
r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]
 
# Make the plot
plt.bar(r1, bars1, color='green', width=barWidth, edgecolor='white', label='# clusters')
plt.bar(r2, bars2, color='gold', width=barWidth, edgecolor='white', label='max set cardinality')
 
# Add xticks on the middle of the group bars
plt.xlabel('(alpha, t) values', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(bars1))], 
           ['(16,20)', '(16,25)', '(16,40)', '(22,40)', '(30,30)', '(30,40)', '(40,40)', '(40,50)', '(40,60)', '(50,40)', '(50,60)'])
 
# Create legend & Show graphic
plt.legend()
plt.show()
```
