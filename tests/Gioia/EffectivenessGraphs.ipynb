## Effectiveness graphical representation

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
import csv

## General data
num_classifiers = 5

centrifuge = []
with open("Centrifuge_effectiveness.csv", 'r') as file:
    csv_file = csv.DictReader(file)
    for row in csv_file:
        if row['Parameter']!= 'classifier':
            centrifuge.append((row['Parameter'], int(row['Value'])))
        else:
            centrifuge.append((row['Parameter'], row['Value']))

kraken1 = []
with open("kraken1_effectiveness.csv", 'r') as file:
    csv_file = csv.DictReader(file)
    for row in csv_file:
        if(row['Parameter']!= 'classifier'):
            kraken1.append((row['Parameter'], int(row['Value'])))
        else:
            kraken1.append((row['Parameter'], row['Value']))

kraken2 = []
with open("kraken2_effectiveness.csv", 'r') as file:
    csv_file = csv.DictReader(file)
    for row in csv_file:
        if(row['Parameter']!= 'classifier'):
            kraken2.append((row['Parameter'], int(row['Value'])))
        else:
            kraken2.append((row['Parameter'], row['Value']))
            
clark_species = []
with open("Clark_species_effectiveness.csv", 'r') as file:
    csv_file = csv.DictReader(file)
    for row in csv_file:
        if row['Parameter']!= 'classifier':
            clark_species.append((row['Parameter'], int(row['Value'])))
        else:
            clark_species.append((row['Parameter'], row['Value']))

clark_genus = []
with open("Clark_genus_effectiveness.csv", 'r') as file: 
    csv_file = csv.DictReader(file)
    for row in csv_file:  
        if row['Parameter']!= 'classifier':
            clark_genus.append((row['Parameter'], int(row['Value'])))     
        else:
            clark_genus.append((row['Parameter'], row['Value']))

print(centrifuge)
print(kraken1)
print(kraken2)
print(clark_species)
print(clark_genus)
```

```python
if ( centrifuge[1][1] != kraken1[1][1] or centrifuge[1][1] != kraken2[1][1] or centrifuge[1][1] != clark_species[1][1] or centrifuge[1][1] != clark_genus[1][1] 
    or kraken1[1][1] != kraken2[1][1] or kraken1[1][1] != clark_species[1][1] or kraken1[1][1] != clark_genus[1][1] 
    or kraken2[1][1] != clark_species[1][1] or kraken2[1][1] != clark_genus[1][1] 
    or clark_species[1][1] != clark_genus[1][1]
    or centrifuge[3][1] != kraken1[3][1] or centrifuge[3][1] != kraken2[3][1] or centrifuge[3][1] != clark_species[3][1] or centrifuge[3][1] != clark_genus[3][1] 
    or kraken1[3][1] != kraken2[3][1] or kraken1[3][1] != clark_species[3][1] or kraken1[3][1] != clark_genus[3][1] 
    or kraken2[3][1] != clark_species[3][1] or kraken2[3][1] != clark_genus[3][1] 
    or clark_species[3][1] != clark_genus[3][1]):
    print("Error in input files")

else:
    num_clusters = centrifuge[1][1]
    num_reads = centrifuge [3][1]
```

```python
# num patological clusters retrieved by 'Effectiveness' notebook
centrifuge_patological_cluster = centrifuge[2][1]
kraken1_patological_cluster = kraken1[2][1]
kraken2_patological_cluster = kraken2[2][1]
clark_species_patological_cluster = clark_species[2][1]
clark_genus_patological_cluster = clark_genus[2][1]

# num clusters where the most frequent class is '0' (see 'Effectiveness' notebook)
centrifuge_zero_cluster = centrifuge[5][1]
kraken1_zero_cluster = kraken1[5][1]
kraken2_zero_cluster  = kraken2[5][1]
clark_species_zero_cluster = clark_species[5][1]
clark_genus_zero_cluster = clark_genus[5][1]

patological_read_centrifuge = centrifuge[4][1]
patological_read_kraken1 = kraken1[4][1]
patological_read_kraken2 = kraken2[4][1]
patological_read_clark_species = clark_species[4][1]
patological_read_clark_genus = clark_genus[4][1]

num_zero_read_centrifuge = centrifuge[6][1]
num_zero_read_kraken1 = kraken1[6][1]
num_zero_read_kraken2 = kraken2[6][1] 
num_zero_read_clark_species = clark_species[6][1]
num_zero_read_clark_genus = clark_genus[6][1]

num_non_zero_read_centrifuge = centrifuge[7][1]
num_non_zero_read_kraken1 = kraken1[7][1]
num_non_zero_read_kraken2 = kraken2[7][1]
num_non_zero_read_clark_species = clark_species[7][1]
num_non_zero_read_clark_genus = clark_genus[7][1]

centrifuge_ties = centrifuge[8][1]
kraken1_ties = kraken1[8][1]
kraken2_ties = kraken2[8][1]
clark_genus_ties = clark_species[8][1]
clark_species_ties = clark_genus[8][1]

centrifuge_zero_ties = centrifuge[9][1]
kraken1_zero_ties = kraken1[9][1]
kraken2_zero_ties = kraken2[9][1]
clark_genus_zero_ties = clark_species[9][1]
clark_species_zero_ties = clark_genus[9][1]
```

```python
labels = ['Centrifuge', 'Kraken1', 'Kraken2', 'Clark Species', 'Clark Genus']

patological_classifier = [centrifuge_patological_cluster, kraken1_patological_cluster, kraken2_patological_cluster,
                          clark_species_patological_cluster, clark_genus_patological_cluster]

width = 0.35       # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots()

ax.bar(labels, patological_classifier, width, label='Patological Clusters', color='springgreen')

plt.ylim(0, max(patological_classifier) + max(patological_classifier)/100)
plt.yticks(np.arange(0, max(patological_classifier) + max(patological_classifier)/100, max(patological_classifier)/10))

fig.set_size_inches(10, 7)

ax.set_ylabel('number of clusters', fontsize = 10)
ax.set_title('Patological Clusters by Classifier', fontsize = 15)
ax.legend(fontsize = 10, loc='upper left' )

plt.show()
```

```python
labels = ['Centrifuge', 'Kraken1', 'Kraken2', 'Clark Species', 'Clark Genus']

patological_classifier = [centrifuge_patological_cluster, kraken1_patological_cluster, kraken2_patological_cluster, 
                          clark_species_patological_cluster, clark_genus_patological_cluster]

cluster_tot = [num_clusters, num_clusters, num_clusters, num_clusters, num_clusters]

width = 0.35       # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots()

plt.ylim(0, max(cluster_tot) + max(cluster_tot)/100)
plt.yticks(np.arange(0, max(cluster_tot) + max(cluster_tot)/100, max(cluster_tot)/10))

ax.bar(labels, cluster_tot, width, label='Total Clusters', color='paleturquoise')
ax.bar(labels, patological_classifier, width, label='Patological Clusters', color='springgreen')

fig.set_size_inches(10, 7)

ax.set_ylabel('number of clusters', fontsize = 10)
ax.set_title('Patological Clusters wrt Total Clusters by Classifier', fontsize = 15)
ax.legend()

plt.show()
```

```python
labels = ['Centrifuge', 'Kraken1', 'Kraken2', 'Clark Species', 'Clark Genus']

zero_classifier = [centrifuge_zero_cluster, kraken1_zero_cluster, kraken2_zero_cluster, clark_species_zero_cluster, clark_genus_zero_cluster]

cluster_tot = [num_clusters, num_clusters, num_clusters, num_clusters, num_clusters]

width = 0.35       # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots()

plt.ylim(0,max(cluster_tot) + max(cluster_tot)/100)
plt.yticks(np.arange(0, max(cluster_tot) + max(cluster_tot)/100, max(cluster_tot)/10))

fig.set_size_inches(10, 7)

ax.bar(labels, cluster_tot, width, label='Total Clusters', color='paleturquoise')
ax.bar(labels, zero_classifier, width, label=' Majority Zero Clusters', color='lightseagreen')

ax.set_ylabel('number of clusters', fontsize = 10)
ax.set_title('Majority Zero Clusters by Classifier', fontsize = 15)
ax.legend(loc = 'upper right')

plt.show()
```

```python
labels = ['Centrifuge', 'Kraken1', 'Kraken2', 'Clark Species', 'Clark Genus']

patological_classifier = [centrifuge_patological_cluster, kraken1_patological_cluster, kraken2_patological_cluster, 
                          clark_species_patological_cluster, clark_genus_patological_cluster]

zero_classifier = [centrifuge_zero_cluster, kraken1_zero_cluster, kraken2_zero_cluster, clark_species_zero_cluster, clark_genus_zero_cluster]

cluster_tot = [num_clusters, num_clusters, num_clusters, num_clusters, num_clusters]

width = 0.35       # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots()

plt.ylim(0,max(cluster_tot) + max(cluster_tot)/100)
plt.yticks(np.arange(0, max(cluster_tot) + max(cluster_tot)/100, max(cluster_tot)/10))

fig.set_size_inches(10, 7)

ax.bar(labels, cluster_tot, width, label='Total Clusters', color='paleturquoise')
ax.bar(labels, patological_classifier, width, label='Patological Clusters', color='springgreen')
ax.bar(labels, zero_classifier, width, bottom=patological_classifier, label=' Majority Zero Clusters', color='lightseagreen')

ax.set_ylabel('number of clusters', fontsize = 10)
ax.set_title('Patological + Majority Zero Clusters by Classifier', fontsize = 15)
ax.legend(loc = 'upper right')

plt.show()
```

```python
num_read_list = []

for i in range(0, num_classifiers):
    num_read_list.append(num_reads)
```

```python
labels = ['Centrifuge', 'Kraken1', 'Kraken2', 'Clark Species', 'Clark Genus']

patological_read = [patological_read_centrifuge, patological_read_kraken1, patological_read_kraken2,
                         patological_read_clark_species, clark_genus_patological_cluster]

width = 0.35       # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots()

ax.bar(labels, patological_read, width, label='Patological Reads', color='hotpink')

plt.ylim(0, max(patological_read) + max(patological_read)/100)
plt.yticks(np.arange(0, max(patological_read) + max(patological_read)/100, max(patological_read)/10))

fig.set_size_inches(10, 7)

ax.set_ylabel('number of reads', fontsize = 10)
ax.set_title('Patological Reads by Classifier', fontsize = 15)
ax.legend()

plt.show()
```

```python
labels = ['Centrifuge', 'Kraken1', 'Kraken2', 'Clark Species', 'Clark Genus']

num_zero_read = [num_zero_read_centrifuge, num_zero_read_kraken1, num_zero_read_kraken2,
                         num_zero_read_clark_species, num_zero_read_clark_genus]

num_non_zero_read = [num_non_zero_read_centrifuge, num_non_zero_read_kraken1, num_non_zero_read_kraken2,
                         num_non_zero_read_clark_species, num_non_zero_read_clark_genus]

width = 0.35       # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots()

ax.bar(labels, num_read_list, width, label='Total Reads', color='lightpink')
ax.bar(labels, num_zero_read, width, label='Zero Reads', color='purple')
ax.bar(labels, num_non_zero_read, width, bottom=num_zero_read, label='Non Zero Read', color='palevioletred')

fig.set_size_inches(10, 7)

plt.ylim(0, num_reads + num_reads/100)
plt.yticks(np.arange(0, num_reads + num_reads/100, num_reads/10))

ax.set_ylabel('number of reads', fontsize = 10)
ax.set_title('Zero and Non Zero Reads by Classifier', fontsize = 15)
ax.legend(loc = 'upper right')

plt.show()
```

```python
labels = ['Centrifuge', 'Kraken1', 'Kraken2', 'Clark Species', 'Clark Genus']

patological_read = [patological_read_centrifuge, patological_read_kraken1, patological_read_kraken2,
                         patological_read_clark_species, clark_genus_patological_cluster]

num_zero_read = [num_zero_read_centrifuge, num_zero_read_kraken1, num_zero_read_kraken2,
                         num_zero_read_clark_species, num_zero_read_clark_genus]

num_non_zero_read = [num_non_zero_read_centrifuge, num_non_zero_read_kraken1, num_non_zero_read_kraken2,
                         num_non_zero_read_clark_species, num_non_zero_read_clark_genus]

sum = []

for i in range(0, num_classifiers):
    sum.append(patological_read[i] + num_zero_read[i])
    

width = 0.35       # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots()

ax.bar(labels, num_read_list, width, label='Total Reads', color='lightpink')
ax.bar(labels, patological_read, width, label='Patological Reads', color='hotpink')
ax.bar(labels, num_zero_read, width, bottom = patological_read, label='Zero Reads', color='purple')
ax.bar(labels, num_non_zero_read, width, bottom=sum, label='Non Zero Read', color='palevioletred')

fig.set_size_inches(10, 7)

plt.ylim(0, num_reads + num_reads/100)
plt.yticks(np.arange(0, num_reads + num_reads/100, num_reads/10))

ax.set_ylabel('number of reads', fontsize = 10)
ax.set_title('Patological + Zero + Non Zero reads by Classifier', fontsize = 15)
ax.legend(loc = 'upper right')

plt.show()
```

```python
labels = ['Centrifuge', 'Kraken1', 'Kraken2', 'Clark Species', 'Clark Genus']

patological_read = [patological_read_centrifuge, patological_read_kraken1, patological_read_kraken2,
                         patological_read_clark_species, clark_genus_patological_cluster]
sum = []

for i in range(0, num_classifiers):
    sum.append(num_zero_read[i] + num_non_zero_read[i] + patological_read[i])
    
width = 0.35       # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots()

ax.bar(labels, num_read_list, width, label='Total Reads', color='lightpink')
ax.bar(labels, sum, width, label='Zero Reads', color='purple')

fig.set_size_inches(10, 7)

plt.ylim(0, num_reads + num_reads/100)
plt.yticks(np.arange(0, num_reads + num_reads/100, num_reads/10))

ax.set_ylabel('number of reads')
ax.set_title('After Total Reassignment', fontsize = 15)
ax.legend()

plt.show()
```
